**Project Name : Exploratory Data Analysis of Amazon Prime Video Content**

**Project Summary:**

The gole of the project is to perform Exploratory Data Analysis(EDA) o the Amazon Prime Video Dataset, focusing on understanding the content cateloge of Tv shows and movies available on the platform. The Analysis uncovers trends, insights and patterns related to content type genre distribution relating release_years, and regional availability.

**Key Steps :**

**Data Collection & Cleaning** - The project used to datasets containing information about titles, and credits for the movies and Tv shows on Amazon Prime Video. The Data was cleaned by handling missing values, removing duplicates and erasing data consistency.

**Data Visualization** - Various visualization techniques were used,including Box Plots, Histogram, Bar Plots, Scatter Plots, Line Plots, Pie charts and violin plots, these visualizations helped uncover patterns and relationships within the data.

**Insights** - The Analysis revealed several key insights:





*    Amazon Prime Videos has more movies than Tv shows. Drama and comedy are the most common genres.
*   Movie run times have slightly decreased over the years, while Tv    shows has more seasons.
*   Older titles tend to have higher rating than never results.
*   The United States is the primary content Producer, Followed by India and the United Kingdom.

**Missing Values/ Null Values**

```python
#check for null values
data.isnull().sum()
```

we have null values in the description column. So lets drop it:

```python
#drop null values in the description column
data.dropna(subset=['description'], inplace=True)
```

```python
#mode of age certification
data['age_certification'].mode()[0]
```

```python
#replace null values in imdb_score with mean of imdb_score
data['age_certification'].fillna(data['age_certification'].mode()[0], inplace = True)
```

```python
data.isnull().sum()
```

The seasons columns has null values because it's only applicable to TV Shows but not for Movies, instead of this removing nulls replace them with '0'

```python
#replace null values in the seasons column with 0
data['seasons'].fillna(0, inplace = True)
```

```python
data.isnull().sum()
```

IMDB_ID is unique for movies and Tv shows, lets drop this entire columns

```python
data.dropna(subset = ['imdb_id'], inplace = True)
```

```python
#replace null values in the seasons column with 0
data['imdb_id'].fillna(0, inplace = True)
```

```python
data.isnull().sum()
```

IMDB_SCORE represents the rating of a movie or Tv show, remove null values and replace them with mean.

```python
round(data['imdb_score'].mean(),1)
```

```python
#replace null values in the imdb score with mean of imdb_score
data['imdb_score'].fillna(round(data['imdb_score'].mean(),1), inplace = True)
```

```python
data.shape
```

IMDB_VOTES represents the no.of votes received bt Tv shows and Movies. Missing null values and replace them with 0

```python
data['imdb_votes'].fillna(0, inplace = True)
```

```python
data.isnull().sum()
```

TMDB_SCORE is similar to IMDB_SCORE,


```python
#mean of tmdb popularity
round(data['tmdb_popularity'].mean(),1)
```

```python
#replace null values in the tmdb_popularity with column mean
data['tmdb_popularity'].fillna(round(data['tmdb_popularity'].mean(),1), inplace = True)
```

```python
#mean of tmdb_score
round(data['tmdb_score'].mean(),1)
```

```python
#replace all null values in tmdb_score with column mean
data['tmdb_score'].fillna(round(data['tmdb_score'].mean(),1), inplace = True)
```

Missing values in the Character column, replace them by "Unknown"

```python
data['character'].fillna('unknown', inplace = True)
```
